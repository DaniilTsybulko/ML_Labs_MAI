# С помощью графика покажите:

```python
import matplotlib.pyplot as plt
```

6. Сколько женщин и мужчин участвовало в Летних Олимпийских играх каждый год?

```python
gender_participation_per_year = df.groupby(['Year', 'Sex']).size().reset_index(name='Count')

plt.figure(figsize=(20, 6))
for sex in gender_participation_per_year['Sex'].unique():
    data = gender_participation_per_year[gender_participation_per_year['Sex'] == sex]
    plt.bar(data['Year'], data['Count'], label=f'{sex}')

plt.xlabel('Год')
plt.ylabel('Количество учасников')
plt.title('Количество участников на Летних Олимпийских игр по году и полу')
plt.legend(title="Пол")
plt.xticks(gender_participation_per_year['Year'].unique(), rotation=45)
plt.show()
```

7. Какие 10 стран на текущий момент имеют наибольшее количество золотых медалей?


```python
gold_country_score = df[df["Medal"] == "Gold"].groupby("NOC").count().sort_values(by="Medal").tail(10)
gold_country_score.iloc[:, -1:]
```

```python
x = gold_country_score.index
y = gold_country_score["Medal_bool"]

plt.bar(x, y)
plt.xlabel('Страна')
plt.ylabel('Количество золотых медалей')
plt.title('Топ-10 стран по количеству золотых медалей')
plt.show()
```

8. Сколько золотых, серебряных и бронзовых медалей зарабатывал СССР отдельно по каждым играм?

```python
gold_ussr_score = df[(df["NOC"] == "URS") & (df["Medal"] == "Gold")].groupby("Year").count()
silver_ussr_score = df[(df["NOC"] == "URS") & (df["Medal"] == "Silver")].groupby("Year").count()
bronze_ussr_score = df[(df["NOC"] == "URS") & (df["Medal"] == "Bronze")].groupby("Year").count()

join_table = gold_ussr_score.iloc[:, :1].join(silver_ussr_score.iloc[:, 1:2]).join(bronze_ussr_score.iloc[:, 2:3])
join_table = join_table.rename(columns={"player_id": "gold_score", "Name": "silver_score", "Sex": "bronze_score"})

join_table
```

```python
x = join_table.index

plt.plot(x, join_table["gold_score"], '-', x, join_table["silver_score"], '-', x, join_table["bronze_score"], '-')
plt.xlabel('Год')
plt.ylabel('Количество медалей')
plt.title('Медали СССР по годам')
plt.xticks(x.unique())
plt.legend(("золотые медали", "серебрянные медали", "бронзовые медали"))
plt.show()
```

9. Сколько медалей завоёвывали спортсмены и спортсменки (отдельно) команды Китая на каждой олимпиаде?


```python
chn_medals_male = df[(df["NOC"] == "CHN") & (df["Sex"] == "M") & (df["Medal_bool"] == 1)].groupby("Year").count()
chn_medals_female = df[(df["NOC"] == "CHN") & (df["Sex"] == "F") & (df["Medal_bool"] == 1)].groupby("Year").count()

join_table_chn = chn_medals_male.iloc[:, :1].join(chn_medals_female.iloc[:, 1:2])
join_table_chn = join_table_chn.rename(columns={"player_id": "chn_medals_male", "Name": "chn_medals_female"})

join_table_chn
```

```python
x = join_table_chn.index

width = 1
fig, ax = plt.subplots()

ax.bar(x - width/2, join_table_chn["chn_medals_male"], width, label='Мужчины')
ax.bar(x + width/2, join_table_chn["chn_medals_female"], width, label='Женщины')


ax.set_ylabel('Количество медалей')
ax.set_title('Количество медалей КНР по году и полу')
plt.xticks(x.unique(), rotation=45)
ax.legend(loc='upper left', title='Пол')
plt.show()
```

10. Какие 10 стран имеют наибольшее количество участников, не получивших никаких медалей ни на одной из Летних олимпиад?

Создадим новую фичу history_medal_count, где у нас будет храниться количество медалей, которое завоевал один спортсмен

```python
df_history_medal_count = df.groupby("Name")["Medal_bool"].sum()
df_history_medal_count = df_history_medal_count.to_frame("history_medal_count")
df_history_medal_count['Name'] = df_history_medal_count.index
df_history_medal_count.index.name = None
df_history_medal_count = df_history_medal_count.reset_index(drop=True)
df_history_medal_count
```

```python
df = pd.merge(df, df_history_medal_count, on='Name', how='left')
df
```

```python
loosers = df[df["history_medal_count"] == 0]
loosers = loosers.groupby("NOC").count().sort_values(by="Medal").tail(10).iloc[:, -1:]
```

```python
loosers
```

Юху, наконец-то график

```python
x = loosers.index
y = loosers["history_medal_count"]

plt.bar(x, y)
plt.xlabel('Страна')
plt.ylabel('Количество людей без медалей')
plt.title('Топ-10 стран, которые имеют наибольшее количество участников, не получивших никаких медалей ни на одной из Летних олимпиад')
plt.show()
```
